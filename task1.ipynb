IRIS_classification dataset
using python,ml libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
columns=['sepal length','sepal width','petal length','petal width','class_labels']
kula_dataframe=pd.read_csv('iris.data',names=columns)
kula_dataframe.head(10)
#iris data set values
```

```python
#all numpy values for dataset
kula_dataframe.describe()
```

seaborn graph for comparision of each columns for the final output(species)

```python
sns.pairplot(kula_dataframe,hue='class_labels')
```

data analysis using seaborn

```python
#individual data for each
sns.set_style("whitegrid")
sns.FacetGrid(kula_dataframe, hue ="class_labels",
              height = 6).map(plt.scatter,
                              'sepal length',
                              'petal length').add_legend()
```

data analysis using histogram

```python
#data analysis using histogram
fig, axes = plt.subplots(2, 2, figsize=(20,20))
 
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(kula_dataframe['sepal length'], bins=7)
 
axes[0,1].set_title("Sepal Width")
axes[0,1].hist(kula_dataframe['sepal width'], bins=5);
 
axes[1,0].set_title("Petal Length")
axes[1,0].hist(kula_dataframe['petal length'], bins=6);
 
axes[1,1].set_title("Petal Width")
axes[1,1].hist(kula_dataframe['petal width'], bins=6);
```

taking all values till 3 rows means petal length,sepal length,sepal width,petal length and the output is species denoted as kula_y 

```python
kula_data=kula_dataframe.values
kula_x=kula_data[:,0:4]
kula_y=kula_data[:,4]
kula_x,kula_y
```

training the dataset for the all lengths and widths to get desired species named y_train and y_test are the outputs for the x_train,x_test
here it will learn from the x_train and y_train then applies for x_test and gives output 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(kula_x,kula_y,test_size=0.2)
print(x_train)
```

y_Train details

```python
print(y_train)
```

****************************************************
SVC training model
we will fit our training model first 
then using this we predict output for x_test
and need to compare this prediction results with y_test and we named this as accuracy_score

```python
from sklearn.svm import SVC
kula_svc=SVC()
kula_svc.fit(x_train,y_train)
kula_svc
prediction=kula_svc.predict(x_test)
from sklearn.metrics import accuracy_score
print("accuracy is")
print(accuracy_score(y_test,prediction)*100)
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

****************************************************
logistic regression training model
we will fit our training model first 
then using this we predict output for x_test
and need to compare this prediction results with y_test and we named this as accuracy_score

```python
from sklearn.linear_model import LogisticRegression
kula_lr=LogisticRegression()
kula_lr.fit(x_train,y_train)
predictionlr=kula_lr.predict(x_test)
from sklearn.metrics import accuracy_score
print("accuracy is")
print(accuracy_score(y_test,predictionlr)*100)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictionlr))
```

****************************************************
decision tree classifier training model
we will fit our training model first 
then using this we predict output for x_test
and need to compare this prediction results with y_test and we named this as accuracy_score

```python
from sklearn.tree import DecisionTreeClassifier
kula_tree=DecisionTreeClassifier()
kula_tree.fit(x_train,y_train)
prediction2=kula_tree.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction2))
```
